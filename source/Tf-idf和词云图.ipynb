```python
import pandas as pd
import datetime
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
print('开始时间：' + str(datetime.datetime.now()))
df = pd.read_excel('data1分词后.xlsx', header=0)
all_words = [] # 创建一个空列表来存储分词结果
for i in range(len(df)):
    content = df.分词后内容[i]
    all_words.extend(content)

seg_str = '' # 创建空的字符串用来存储分词结果
for word in all_words:
    seg_str += word


vectorizer = TfidfVectorizer()
 # 转成矩阵
tfidf_matrix = vectorizer.fit_transform([seg_str])
df = pd.DataFrame(
    # 转成数组
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)
# 排序 + 筛选
top_indices = df.iloc[0].argsort()[-20:][::-1].values
top_words = [
    vectorizer.get_feature_names_out()[index]
    for index in top_indices
]
print(f"Importance_words: {','.join(top_words)}")


wordcloud = WordCloud(font_path='微软雅黑.ttf', width=800, height=400, background_color='white').generate(' '.join(all_words))
print("第一首歌词云图")
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
print('\n结束时间：' + str(datetime.datetime.now()))
```
